```python
import pandas as pd

k_air = pd.DataFrame(pd.read_csv('krk-airlines.csv'))
k_fly = pd.DataFrame(pd.read_csv('krk-flights.csv'))
k_pass = pd.DataFrame(pd.read_csv('krk-passengers.csv'))
k_air
```

```python
flight_destination = k_fly.loc[:,['flight','to']]
flight_destination
```

```python
aircraft= k_air.loc[:,["airlineid","airplane"]]
pd.merge(k_fly,aircraft, on='airlineid')
```

```python
pd.merge(k_fly.loc[:,["flight","to"]].query("to == 'London'"),k_pass.loc[:,["flight","name","surname"]],on="flight").sort_values(by="surname")
```

```python
pd.merge(k_fly.loc[:,['flight','to']] , k_pass.query('gender=="Female"').loc[:,["flight","name","surname"]], on="flight").sort_values(by="surname")
```
